## Quick Time Series Analysis

Import libraries of interest

```python
import numpy as pn
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

import os
import time
```

Import the dataset

```python
# Dataset obtained from Google Trends
df = pd.read_csv("got_searches_dataset.csv")
```

Check your data

```python
df.head()
```

Define the columns of interest

```python
df.columns = ['Month','GOT','HBO']
```

```python
df.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

Plot the raw data

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)

plt.xlabel('Month', fontsize=20);
plt.ylim(0, 150)
```

Define the searchers as lists

```python
GOT = df[['GOT']]
HBO = df[['HBO']]

X=2
```

Use a rolling average to smooth your data and eliminate noise, and then plot it

```python
df_rm = pd.concat([GOT.rolling(X).mean(), HBO.rolling(X).mean()], axis = 1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.title("Google Search Popularity of 'Game of Thrones' and 'HBO' from 2010 to 2019", fontsize=20)

plt.ylabel('Popularity (%)', fontsize=20)
plt.ylim(0, 120)

plt.grid(b=None, which='minor', axis='both')

# Season 1
#plt.axvline('2011-04-17', color='r', linestyle='--', lw=2)
plt.text(495,25,'Season 1',rotation=45)

# Season 2
plt.text(505,30,'Season 2',rotation=45)

# Season 3
plt.text(518,38,'Season 3',rotation=45)

# Season 4
plt.text(530,50,'Season 4',rotation=45)

# Season 5
plt.text(542,43,'Season 5',rotation=45)

# Season 6
plt.text(554,58,'Season 6',rotation=45)

# Season 7
plt.text(568,80,'Season 7',rotation=45)

# Season 8
plt.text(586,104,'Season 8 so far',rotation=45)

plt.savefig('GoT_timeseries.png')
```
